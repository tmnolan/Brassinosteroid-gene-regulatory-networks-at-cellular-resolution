# Cell Oracle BaseGRNs

```r
library(tidyverse)
library(ComplexHeatmap)
library(circlize)
library(GeneOverlap)
```

### CellOracle inputs can be downloaded from [ARVEX](https://shiny.mdc-berlin.de/ARVEX/)


```r
(ATAC_base_grn <- read_csv("./data/CellOracle_Input/scATAC_base_GRN.csv") %>%
select(TF, Target_genes) %>%
mutate(Source="scATAC"))
```

```r
(DAPseq_base_grn <- read_csv("./data/CellOracle_Input/DAPseq_GRN.csv") %>%
select(TF, Target_genes) %>%
mutate(Source="DAPseq"))
```

```r
(iGRN_base_grn <- read_csv("./data/CellOracle_Input/iGRN.csv") %>%
select(TF, Target_genes) %>%
mutate(Source="iGRN"))
```

```r
(gene_names <- read.csv("./data/features.tsv.gz", sep="\t", header=FALSE, col.names = c("GeneID", "Name", "V3")) %>%
select(-V3))
```

```r
# combine all subnetworks
base_grn <- bind_rows(ATAC_base_grn, iGRN_base_grn, DAPseq_base_grn) %>%
filter(TF %in% gene_names$GeneID) %>%
filter(Target_genes %in% gene_names$GeneID)
```

```r
# pre combined base GRN

combined_base_grn <- read_csv("./data/CellOracle_Input/Combined_base_GRN.csv") %>%
select(TF, Target_genes) %>%
mutate(Source="All")
```

```r
#combine all with subcatagories for stat and only retain genes that we map in single-cell

total_base_grn <- bind_rows(combined_base_grn, base_grn) %>%
filter(TF %in% gene_names$GeneID) %>%
filter(Target_genes %in% gene_names$GeneID) %>% 
mutate(TF_target=paste(TF, Target_genes, sep="_"))
```

```r
(combined_stats <- total_base_grn %>%
  group_by(Source) %>%
  summarize(TFs = n_distinct(TF), 
           Targets=n_distinct(Target_genes), 
           Edges = n_distinct(TF_target)))
```

```r
# check overlap in GRNs
for_olap <- filter(total_base_grn, Source!="All") %>%
select(Source, TF_target)
```

```r
# get background of all possible TF-target combos 

All_TFs <- total_base_grn %>%
select(TF) %>%
distinct()
```

```r
All_targets <- total_base_grn %>%
select(Target_genes) %>%
distinct()
```

```r
All_TF_target_combos <- 1601 * 31019
```

```r
options(repr.plot.width = 14, repr.plot.height = 12)


markers_list <- split(for_olap, f=for_olap$Source)
 
#this makes list from long df of gene lists - TARGET is what we want to keep
markers_list  <- lapply(markers_list, function(x) x[names(x)=="TF_target"])
# convert each sublist into character and eliminate duplicates
markers_list  <- lapply(markers_list, function(x) as.character(unique(x$TF_target)))
 
## GeneOverlap 

# nfeatures RNA

genome_size <- All_TF_target_combos

#compare all lists
gom.self <- newGOM(markers_list, markers_list, genome.size=genome_size)


int <- getNestedList(gom.self, "intersection")
int_matrix <- getMatrix(gom.self, "intersection")
p.val <- getMatrix(gom.self, "pval")
JC <- getMatrix(gom.self, "Jaccard")

# log of p.val for intersection
p.val_log <- -log10(p.val + 1e-200)
                        
```

```r
options(repr.plot.width = 8, repr.plot.height = 8)

olap <- Heatmap(p.val_log, name = "-log10_pval", col = colorRamp2(c(0, 200), c("beige", "red")), 
                column_title = "Number of shared edges", cluster_rows = T,cluster_columns = T, use_raster= FALSE, show_column_names = TRUE, show_row_names = TRUE, show_row_dend = TRUE, show_column_dend = TRUE, cell_fun = function(j, i, x, y, width, height, fill) {grid.text(sprintf("%.0f", int_matrix[i, j]), x, y, gp = gpar(fontsize = 10))
}) 
                        
                        # padding - bottom, left, top, right
draw(olap, padding = unit(c(5, 5, 5, 10), "mm"), heatmap_legend_side = "left")
```

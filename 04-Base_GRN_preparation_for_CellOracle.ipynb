```python
# Load scATAC_base_GRN.csv (Downloaded from ARVEX)
base_GRN = pd.read_csv("../Root_scATAC/cicero_output/scATAC_base_GRN.csv", index_col=0)

# Check data
base_GRN.head()
```

```python
d = dict()
for i in range(2,1067):
        d[base_GRN.columns[i]] = list(base_GRN["gene_short_name"].iloc[np.where(base_GRN.iloc[:,i]==1.0)])
```

```python
df = pd.DataFrame()
for i in d.keys():
    df2 = pd.DataFrame()
    df2["TF"]=[i]*len(d[i])
    df2["Target_genes"]=d[i]
    df = df.append(df2)    
```

### CellOracle inputs can be downloaded from [ARVEX](https://shiny.mdc-berlin.de/ARVEX/)


```r
Atrichoblast <- read_csv("./data/WT_CellOracle_full_networks/atrichoblast_Root_Atlas_SCT_celloracle_network_iGRN.csv") %>%
  mutate(network="Atrichoblast")

Columella <- read_csv("./data/WT_CellOracle_full_networks/columella_Root_Atlas_SCT_celloracle_network_iGRN.csv") %>%
  mutate(network="Columella")

Cortex <- read_csv("./data/WT_CellOracle_full_networks/cortex_Root_Atlas_SCT_celloracle_network_iGRN.csv") %>%
  mutate(network="Cortex")

Endodermis <- read_csv("./data/WT_CellOracle_full_networks/endodermis_Root_Atlas_SCT_celloracle_network_iGRN.csv") %>%
  mutate(network="Endodermis")

LateralRootCap <- read_csv("./data/WT_CellOracle_full_networks/lrc_Root_Atlas_SCT_celloracle_network_iGRN.csv") %>%
  mutate(network="LateralRootCap")

Pericycle <- read_csv("./data/WT_CellOracle_full_networks/pericycle_Root_Atlas_SCT_celloracle_network_iGRN.csv") %>%
  mutate(network="Pericycle")

Phloem <- read_csv("./data/WT_CellOracle_full_networks/phloem_Root_Atlas_SCT_celloracle_network_iGRN.csv") %>%
  mutate(network="Phloem")

Procambium <- read_csv("./data/WT_CellOracle_full_networks/procambium_Root_Atlas_SCT_celloracle_network_iGRN.csv") %>%
  mutate(network="Procambium")

Trichoblast <- read_csv("./data/WT_CellOracle_full_networks/trichoblast_Root_Atlas_SCT_celloracle_network_iGRN.csv") %>%
  mutate(network="Trichoblast")

Xylem <- read_csv("./data/WT_CellOracle_full_networks/xylem_Root_Atlas_SCT_celloracle_network_iGRN.csv") %>%
  mutate(network="Xylem")
```

```r
Full_net <- bind_rows(Atrichoblast, 
                      Columella, 
                      Cortex, 
                      Endodermis, 
                      LateralRootCap, 
                      Pericycle, 
                      Phloem, 
                      Procambium, 
                      Trichoblast, 
                      Xylem) %>%
select(-X1) %>%
mutate(TF_target=paste(source, target, sep="_"))
```

```r
Full_net
```

```r
(combined_stats <- Full_net %>%
  summarize(TFs = n_distinct(source), 
           Targets=n_distinct(target), 
           Edges = n_distinct(TF_target), 
           GRN_Configurations=n_distinct(cluster)))
```

```r
write_csv(Full_net, "./output/WT_CellOracle_network.csv")
```
